```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
data=pd.read_csv('book.csv',encoding = "ISO-8859-1")
data1=data.rename({'User.ID':'user_id','Book.Title':'book_title','Book.Rating':'book_rating'},axis=1)
data1
```

```python
data2=data1.iloc[:,1:]
data2
```

## EDA

```python
unique_users=len(data2.user_id.unique())
unique_books=len(data2.book_title.unique())
print('Number of unique users: ',unique_users)
print('Number of unique books: ',unique_books)
```

```python
data2.book_rating.value_counts()
```

## Pivot Table

```python
books_pivot= data2.pivot_table(index='user_id',columns='book_title',values='book_rating').reset_index(drop=True)
books_pivot=books_pivot.fillna(0)
```

```python
books_pivot.index=data2.user_id.unique()
books_pivot
```

## Cosine distance

```python
user_sim = 1 - pairwise_distances( books_pivot.values,metric='cosine')
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df.index=data2.user_id.unique()
user_sim_df.columns=data2.user_id.unique()
```

```python
user_sim_df
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
user_sim_df.idxmax(axis=1)
```

```python
flu_user_rating=books_pivot['Classical Mythology']
```

```python
sim_to_flu=books_pivot.corrwith(flu_user_rating)
sim_to_flu.head()
```
